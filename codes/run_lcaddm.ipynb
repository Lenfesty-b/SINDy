# Leaky competing accumulator drift diffusion model 

### Run single trials first
### Using single trial coefs to run average SINDy model

```python
import numpy as np
from lcaddm_st import simulate_trial

# Define parameters for the trial
t = np.arange(0, 100, 0.01)
dt = 0.01
S1 = 3.5
S2 = 3
b = 10
k = 10
c = 0.11
z = 1

# Simulate the trial
x, y, decision_time, choice = simulate_trial(t, dt, S1, S2, b, k, c, z)

# Print the results
print("Decision time:", decision_time)
print("Choice:", "x" if choice == 1 else ("y" if choice == 0 else "No decision"))

# Plot the results
import matplotlib.pyplot as plt

plt.plot(t[:len(x)], x, label='x')
plt.plot(t[:len(y)], y, label='y')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
```

```python
import numpy as np
from lcaddm_st import sessionLCADDM

# Define parameters for the session
trials = 100
signal = 0.5
seeding = 0  # Example seed for reproducibility
z=1 
t = np.arange(0, 10000, 0.01)

# Run the session
coef_mat, choice_trials, sindy_choice, decision_time, sindy_dt, model_data, sindy_data = sessionLCADDM(trials, signal, seeding)

# Print a summary of the results
print("Number of trials:", trials)
print("Average decision time (LCA-DDM):", np.mean(decision_time))
print("Average decision time (SINDy):", np.mean(sindy_dt))
print("LCA-DDM choices (1 for x, 0 for y):", choice_trials)
print("SINDy choices (1 for x, 0 for y):", sindy_choice)

# Plot a sample trial from LCA-DDM and SINDy data
import matplotlib.pyplot as plt

# Plot LCA-DDM sample
sample_x, sample_y = model_data[0]
plt.plot(t[0:len(sample_x)],sample_x, label='x (LCA-DDM)')
plt.plot(t[0:len(sample_y)],sample_y, label='y (LCA-DDM)')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title('Sample LCA-DDM Trial')
plt.legend()
plt.show()

# Plot SINDy sample
sample_sindy_x, sample_sindy_y = sindy_data[0]
plt.plot(t[0:len(sample_sindy_x)],sample_sindy_x, label='x (SINDy)')
plt.plot(t[0:len(sample_sindy_y)],sample_sindy_y, label='y (SINDy)')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title('Sample SINDy Trial')
plt.legend()
plt.show()
```
