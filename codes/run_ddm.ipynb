# Drift diffusion model

### Run single trials first
### Using single trial coefs to run average SINDy model

```python
import numpy as np
from ddm_ST import simulate_trial

# Define parameters for the trial
t = np.arange(0, 100, 0.01)
dt = 0.01
A = 0.5
c = 0.11
z = 1

# Simulate the trial
x, decision_time, choice = simulate_trial(t, dt, A, c, z)

# Print the results
print("Decision time:", decision_time)
print("Choice:", "Correct" if choice == 1 else ("Incorrect" if choice == 0 else "No decision"))

# Plot the results
import matplotlib.pyplot as plt

plt.plot(t[:len(x)], x, label='Decision Variable')
plt.axhline(z, color='r', linestyle='--', label='Correct')
plt.axhline(-z, color='b', linestyle='--', label='Incorrect')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import numpy as np
from ddm_single_trial import sessionDDM

# Define parameters for the session
trials = 100
signal = 0.01
z=1 
t = np.arange(0, 10000, 0.1)
z=1
# Run the session
coef_mat, choice_trials, sindy_choice, decision_time, sindy_dt, model_data, sindy_data = sessionDDM(trials, signal)

# Print a summary of the results
print("Number of trials:", trials)
print("Average decision time (DDM):", np.mean(decision_time))
print("Average decision time (SINDy):", np.mean(sindy_dt))
print("DDM choices (1 for Correct, 0 for Incorrect):", choice_trials)
print("SINDy choices (1 for Correct, 0 for Incorrect):", sindy_choice)

# Plot a sample trial from DDM and SINDy data
import matplotlib.pyplot as plt

# Plot DDM sample
sample_x = model_data[0]
plt.plot(t[0:len(sample_x)],sample_x, label='Decision Variable (DDM)')
plt.axhline(z, color='r', linestyle='--', label='Correct')
plt.axhline(-z, color='b', linestyle='--', label='Incorrect')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title('Sample DDM Trial')
plt.legend()
plt.show()

# Plot SINDy sample
sample_sindy_x = sindy_data[0]
plt.plot(t[0:len(sample_sindy_x)],sample_sindy_x, label='Decision Variable (SINDy)')
plt.axhline(z, color='r', linestyle='--', label='Correct')
plt.axhline(-z, color='b', linestyle='--', label='Incorrect')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title('Sample SINDy Trial')
plt.legend()
plt.show()
```
