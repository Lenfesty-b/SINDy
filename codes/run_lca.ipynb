# Linear leaky competing accumulator 

### Run single trials first
### Using single trial coefs to run average SINDy model

```python
###Single trial

import numpy as np
from lca_ST import simulate_trial

# Define parameters for the trial
t = np.arange(0, 100, 0.01)
dt = 0.01
S1 = 2.85
S2 = 1.85
b = 4
k = 3
c = 0.11
z = 1

# Simulate the trial
x, y, decision_time, choice = simulate_trial(t, dt, S1, S2, b, k, c, z)

# Print the results
print("Decision time:", decision_time)
print("Choice:", "x" if choice == 1 else "y")

# Plot the results
import matplotlib.pyplot as plt

plt.plot(t[:len(x)], x, label='x')
plt.plot(t[:len(y)], y, label='y')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Single trial fitting
import numpy as np
from lca_ST import sessionLCA

# Define parameters for the session
trials = 100
signal = 0
z=1 
t = np.arange(0, 10000, 0.01)

# Run the session
coef_mat, choice_trials, sindy_choice, decision_time, sindy_dt, model_data, sindy_data = sessionLCA(trials, signal)

# Print a summary of the results
print("Number of trials:", trials)
print("Average decision time (LCA):", np.mean(decision_time))
print("Average decision time (SINDy):", np.mean(sindy_dt))
print("LCA choices (0 for y, 1 for x):", choice_trials)
print("SINDy choices (0 for y, 1 for x):", sindy_choice)

# Plot a sample trial from LCA and SINDy data
import matplotlib.pyplot as plt

# Plot LCA sample
sample_x, sample_y = model_data[0]
plt.plot(t[0:len(sample_x)],sample_x, label='x (LCA)')
plt.plot(t[0:len(sample_y)],sample_y, label='y (LCA)')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time(a.u)')
plt.ylabel('Value')
plt.title('Sample LCA Trial')
plt.legend()
plt.show()

# Plot SINDy sample
sample_sindy_x, sample_sindy_y = sindy_data[0]
plt.plot(t[0:len(sample_sindy_x)],sample_sindy_x, label='x (SINDy)')
plt.plot(t[0:len(sample_sindy_y)],sample_sindy_y, label='y (SINDy)')
plt.axhline(z, color='r', linestyle='--', label='Threshold')
plt.xlabel('Time step')
plt.ylabel('Value')
plt.title('Sample SINDy Trial')
plt.legend()
```
